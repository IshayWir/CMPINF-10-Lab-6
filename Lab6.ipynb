```python
# Something™ Program

# Take a name input
name = input("Enter your name: ")

# Take some other input (a number)
number = float(input("Enter a number: "))

# Do Something™ with the other input
# (calculate the factorial if it's a whole number,
# otherwise square it)

if number.is_integer() and number >= 0:
    factorial = 1
    for i in range(1, int(number) + 1):
        factorial *= i
    result = factorial
    action = "the factorial"
else:
    result = number ** 2
    action = "the square"

# Output the result
print(f"\nHello, {name}!")
print(f"You entered {number}.")
print(f"Something™ happened... {action} is {result}!")
```
